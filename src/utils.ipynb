```python
import os
import json

def consolidate_and_sort_scenarios(folder_path, output_filename):
    combined_data = []

    # Check if the directory exists
    if not os.path.exists(folder_path):
        print(f"Error: The folder '{folder_path}' does not exist.")
        return

    # 1. Read and Extract Data
    for filename in os.listdir(folder_path):
        if filename.endswith(".json"):
            file_path = os.path.join(folder_path, filename)
            
            try:
                with open(file_path, 'r', encoding='utf-8') as f:
                    data = json.load(f)
                    
                    for scenario in data:
                        filtered_scenario = {
                            "scenario_id": scenario.get("scenario_id", ""),
                            "scenario_description": scenario.get("scenario_description", "")
                        }
                        combined_data.append(filtered_scenario)
            except Exception as e:
                print(f"Failed to process {filename}: {e}")

    # 2. Sort the list by scenario_id
    # This sorts alphabetically, which works for 'SCEN-001', 'SCEN-002', etc.
    combined_data.sort(key=lambda x: x["scenario_id"])

    # 3. Write to output file
    with open(output_filename, 'w', encoding='utf-8') as f:
        json.dump(combined_data, f, indent=2)
    
    print(f"Success! Processed and sorted {len(combined_data)} scenarios into {output_filename}")

# Run the script
consolidate_and_sort_scenarios("../dataset/scenarios", "llm_experiments/all_scenarios.json")
```
